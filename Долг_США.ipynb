```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Данные о государственном долге
debt_data = [
    {"year": '2018', "debt": 21.516},
    {"year": '2019', "debt": 22.719},
    {"year": '2020', "debt": 26.945},
    {"year": '2021', "debt": 29.617},
    {"year": '2022', "debt": 30.824},
    {"year": '2023', "debt": 31.578},
    {"year": '2024', "debt": 34.001},
]

# Данные о держателях долга
holders_data = {
    "Внутренние держатели": [
        {"name": 'ФРС США', "value": 7.032},
        {"name": 'Пенсионные фонды', "value": 4.821},
        {"name": 'Взаимные фонды', "value": 3.562},
        {"name": 'Банки и страховые компании', "value": 3.241},
        {"name": 'Частные инвесторы', "value": 3.715},
        {"name": 'Прочие внутренние', "value": 2.500}
    ],
    "Иностранные держатели": [
        {"name": 'Япония', "value": 1.105},
        {"name": 'Китай', "value": 0.835},
        {"name": 'Великобритания', "value": 0.715},
        {"name": 'Прочие страны', "value": 6.475}
    ]
}

# График динамики общего долга
years = [data["year"] for data in debt_data]
debt_values = [data["debt"] for data in debt_data]

fig, ax = plt.subplots()
ax.plot(years, debt_values, marker='o', linestyle='-', color='blue')
ax.set_xlabel('Год')
ax.set_ylabel('Триллионы долларов США')
ax.set_title('Динамика общего долга (в триллионах долларов)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# График распределения по держателям долга
fig = make_subplots(rows=1, cols=2, subplot_titles=("Внутренние держатели", "Иностранные держатели"))

for holder_type, holder_details in holders_data.items():
    values = [detail["value"] for detail in holder_details]
    names = [detail["name"] for detail in holder_details]

    fig.add_trace(
        go.Bar(x=values, y=names, orientation='h', name=holder_type),
        row=1, col=1 if holder_type == "Внутренние держатели" else 2
    )

fig.update_layout(title_text="Распределение долга по держателям (начало 2024 года)", barmode='stack', height=600)
fig.show()
```

```python
# Установка необходимых библиотек
!pip install matplotlib plotly

import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Данные о государственном долге
debt_data = [
    {"year": '2018', "debt": 21.516},
    {"year": '2019', "debt": 22.719},
    {"year": '2020', "debt": 26.945},
    {"year": '2021', "debt": 29.617},
    {"year": '2022', "debt": 30.824},
    {"year": '2023', "debt": 31.578},
    {"year": '2024', "debt": 34.001},
]

# Общий долг на 2024 год
total_debt_2024 = 34.001

# Данные о держателях долга
holders_data = {
    "Внутренние держатели": [
        {"name": 'ФРС США', "value": 7.032},
        {"name": 'Пенсионные фонды', "value": 4.821},
        {"name": 'Взаимные фонды', "value": 3.562},
        {"name": 'Банки и страховые компании', "value": 3.241},
        {"name": 'Частные инвесторы', "value": 3.715},
        {"name": 'Прочие внутренние', "value": 2.500}
    ],
    "Иностранные держатели": [
        {"name": 'Япония', "value": 1.105},
        {"name": 'Китай', "value": 0.835},
        {"name": 'Великобритания', "value": 0.715},
        {"name": 'Прочие страны', "value": 6.475}
    ]
}

# Добавление процента к каждому держателю долга
for holder_type in holders_data:
    for holder in holders_data[holder_type]:
        holder["percentage"] = (holder["value"] / total_debt_2024) * 100

# График динамики общего долга
years = [data["year"] for data in debt_data]
debt_values = [data["debt"] for data in debt_data]

fig, ax = plt.subplots()
ax.plot(years, debt_values, marker='o', linestyle='-', color='blue')
ax.set_xlabel('Год')
ax.set_ylabel('Триллионы долларов США')
ax.set_title('Динамика общего долга (в триллионах долларов)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# График распределения по держателям долга
fig = make_subplots(rows=1, cols=2, subplot_titles=("Внутренние держатели", "Иностранные держатели"))

for holder_type, holder_details in holders_data.items():
    values = [detail["value"] for detail in holder_details]
    names = [detail["name"] + f" ({detail['percentage']:.1f}%)" for detail in holder_details]

    fig.add_trace(
        go.Bar(x=values, y=names, orientation='h', name=holder_type),
        row=1, col=1 if holder_type == "Внутренние держатели" else 2
    )

fig.update_layout(title_text="Распределение долга по держателям (начало 2024 года)", barmode='stack', height=600)
fig.show()
```
